```python
#loading XGBoost
import xgboost as xgb
```

```python
from sklearn.model_selection import train_test_split
```

```python
X= work3.drop(columns=['target1', 'target2', 'target3', 'target4'], axis=1)

y= work3['target1']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=3)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
#instantiating xgboost 
xg_clf = xgb.XGBClassifier(booster = 'gblinear', verbosity=3, validate_parameters=True, n_jobs=-1)
```

```python
#creating a param grid for Gridsearch
xg_clf= xgb.XGBClassifier()
param_grid = {
    'eta': [0.1, .25, .05],
    'feature_selector': ['cyclic', 'thrifty', 'greedy', 'shuffle'],
    'reg_lambda':[.5,1.0, 5, 10],
    'reg_alpha':[.5, 1, 5, 10, .25]
    }
```

```python
#gridsearch with XGBoost
grid_clf = GridSearchCV(xg_clf, param_grid, cv=3, scoring='neg_mean_squared_error', n_jobs=-1)

```

```python
grid_clf.fit(X_train, y_train)
```

```python
y_pred=grid_clf.predict(X_test)
```
